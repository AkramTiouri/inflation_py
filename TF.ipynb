```python
def calculer_inflation(ipc_list):
    taux_inflation = []
    for i in range(1, len(ipc_list)):
        ipc_actuel = ipc_list[i]
        ipc_precedent = ipc_list[i - 1]  
        taux_mensuel = ((ipc_actuel - ipc_precedent) / ipc_precedent) * 100
        taux_inflation.append(taux_mensuel)
    
    return taux_inflation

taux_inflation_mensuels = calculer_inflation(values_)
taux_inflation_mensuels.insert(0,0)
plt.figure(figsize=(20,5))
plt.xticks(rotation=90)
plt.grid(True)
plt.plot(dates,taux_inflation_mensuels,marker='o')
plot_filename = "tf.png"
plt.savefig(plot_filename)
plt.show()


        
```

```python
from pandas import read_csv
from pandas import datetime
from pandas import Series
from matplotlib import pyplot
def difference(dataset, interval=1):
     diff = list()
     for i in range(interval, len(dataset)):
            value = dataset[i] - dataset[i - interval]
            diff.append(value)
     return Series(diff)
 
series = Tf["Taux d'inflation"]
X = series.values
diff = difference(X)
```

```python
result = adfuller(diff)
 
print('Statistiques ADF : {}'.format(result[0]))
print('p-value : {}'.format(result[1]))
print('Valeurs Critiques :')
for key, value in result[4].items():
    print('\t{}: {}'.format(key, value))
```
